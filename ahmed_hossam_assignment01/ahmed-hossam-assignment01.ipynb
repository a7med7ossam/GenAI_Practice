```python
# Use pipeline as a high-level
from transformers import pipeline

llm = pipeline("text-generation", model=("LiquidAI/LFM2-700M"))

prompt = """ 
you are an expert chef.
questions : What are the ingredients of 'Burrata Pugliese con Prosciutto di Parma e Fichi Caramellati'?
"""

response = llm(prompt)
```

```python
generated_text = response[0]["generated_text"]
clean_output = generated_text.replace(prompt, "").strip()
print(clean_output)
```
